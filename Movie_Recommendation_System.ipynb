```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
df["features"] = df["genre"] + " " + df["description"]
df[["title", "features"]].head()
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["features"])
```

```python
similarity = cosine_similarity(count_matrix)
similarity
```

```python
def recommend_movie(movie_name):
    movie_name = movie_name.lower()
    
    matching_movies = df[df["title"].str.lower() == movie_name]
    
    if matching_movies.empty:
        return "Movie not found in dataset."
    
    movie_index = matching_movies.index[0]
    similarity_scores = list(enumerate(similarity[movie_index]))
    sorted_movies = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    recommendations = []
    for movie in sorted_movies[1:6]:
        recommendations.append(df.iloc[movie[0]]["title"])
    
    return recommendations
```

```python
recommend_movie("Avatar")
```

```python
recommend_movie("Titanic")
```

```python
recommend_movie("Frozen")
```
